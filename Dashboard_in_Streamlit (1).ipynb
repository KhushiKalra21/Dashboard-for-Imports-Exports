### About the Dataset:

The dataset represents trade transactions, providing essential details about various products being imported and exported between countries. It is designed to assist stakeholders in understanding trade patterns, product demand, and supply chain management. This dataset includes key attributes such as transaction ID, product category, country, port of entry or exit, shipping method, and payment terms. It also captures quantitative details such as quantity, value, and weight of shipments, allowing for a comprehensive analysis of trade activities.


### Aim:
To assist companies, logistics managers, and trade analysts in making informed decisions regarding international trade by understanding the flow of products, shipping methods, and financial details.

### Key Attributes in the Dataset:

- **Transaction_ID**: A unique identifier for each transaction.
- **Country**: The country involved in the import/export transaction.
- **Product**: The type of product involved in the trade.
- **Import_Export**: Indicates whether the transaction is an import or export.
- **Quantity**: The number of units of the product in the transaction.
- **Value**: The monetary value of the transaction in local currency.
- **Date**: The date the transaction took place.
- **Category**: The category of the product (e.g., Clothing, Machinery, Electronics, etc.).
- **Port**: The port where the transaction was processed.
- **Customs_Code**: The customs classification code of the product.
- **Weight**: The weight of the shipment in kilograms.
- **Shipping_Method**: The method used for shipping (e.g., Air, Sea, etc.).
- **Supplier**: The company or entity supplying the product.
- **Customer**: The entity purchasing or receiving the product.
- **Invoice_Number**: The invoice number for the transaction.
- **Payment_Terms**: The agreed payment terms (e.g., Prepaid, Cash on Delivery, etc.).

### About the Dashboard: 
The dashboard consists of 6 interactive charts that provide key insights into international trade. Each chart uses a different variable to offer a comprehensive view of trade activities. The visualizations include a Treemap Chart, two Sunburst Charts, a Bubble Chart, and vertical and horizontal bar charts, all of which offer interactivity through dropdowns or checklists. 

#### Each chart in the dashboard represents:

**Treemap Chart** – 
- **Title**: “Product Category vs Country (Import/Export)”
- **Variables Used**: Country, Import/Export status, Product Category.
- **Observation**: This chart provides a visual representation of product categories traded between various countries. The size of each block represents the volume of trade for a particular product category, offering insights into which countries are importing or exporting specific product types. For example, countries like China might have a significant portion of Electronics exports, while others may dominate in Machinery imports.

**1st Sunburst Chart** – 
- **Title**: “Port vs Shipping Method for Product Categories”
- **Variables Used**: Port, Shipping Method, Product Category.
- **Observation**: This chart displays the distribution of different product categories across various ports and shipping methods (e.g., air, sea). It helps logistics managers identify which shipping methods are predominantly used for certain products and ports. For example, the chart might show that air transport is primarily used for Electronics from specific ports, while sea transport dominates heavy goods like Machinery.

**Bubble Chart** – 
- **Title**: “Quantity vs Value of Trade Transactions”
- **Variables Used**: Product Category, Quantity, Value.
- **Observation**: The Bubble Chart illustrates the relationship between the quantity of products traded and the monetary value of these transactions. Each bubble represents a product category, with the size of the bubble corresponding to the value of the trade. This chart allows stakeholders to understand the value generated per unit quantity across product categories. It’s useful in identifying high-value, low-volume products like Electronics, as opposed to high-volume, low-value products like Clothing.

**Vertical Bar Chart** – 
- **Title**: “Trade Volume by Port (with Shipping Method Differentiation)”
- **Variables Used**: Port, Shipping Method, Quantity.
- **Observation**: This chart showcases the trade volume processed through different ports, with a differentiation based on shipping method (air, sea, etc.). It helps identify the most active ports and the preferred shipping methods for each port, which is valuable for optimizing logistics and managing port capacity.

**2nd Sunburst Chart** – 
- **Title**: “Country-wise Product Categories and Payment Terms”
- **Variables Used**: Country, Product Category, Payment Terms.
- **Observation**: This chart visualizes the distribution of payment terms across countries and product categories. It helps trade managers understand the preferred financial arrangements (e.g., Prepaid, Net 60, Cash on Delivery) by different countries and products. For instance, certain countries may predominantly opt for 'Prepaid' arrangements for electronics but 'Net 60' terms for clothing, influencing trade negotiation strategies.

**Horizontal Bar Chart** – 
- **Title**: “Country-wise Customs Code Analysis for Product Categories”
- **Variables Used**: Country, Customs Code, Product Category.
- **Observation**: The horizontal bar chart shows the breakdown of customs codes by country and product category, allowing companies to see which products face significant regulatory scrutiny. It also aids in identifying trends in customs classifications, helping companies streamline documentation and reduce delays.
